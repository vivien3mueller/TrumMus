```python
from flexiconc import Concordance
c = Concordance()
```

```python
c.retrieve_from_clic(api_base_url="https://clic-fiction.com/api/concordance", query=["stomach"], corpora="dickens")
```

```python
c.root
```

```python
from IPython.display import HTML, display
from flexiconc.visualization.html_visualizer import generate_concordance_html, generate_analysis_tree_html
display(HTML(generate_analysis_tree_html(c)))
```

```python
display(HTML(generate_concordance_html(c, c.root, n=1000)))
```

```python
algos = c.root.available_algorithms()
sorted(list(algos.keys()))
```

```python
n1 = c.root.add_subset_node(
    ("Random Sample",
     {'sample_size': 10, 'seed': 9101989})
)
display(HTML(generate_concordance_html(c, n1, n=1000)))
```

```python
display(HTML(generate_analysis_tree_html(c)))
```

```python
c.root.schema_for("Random Sample")
```

```python
display(HTML(generate_concordance_html(c, n1, n=1000, metadata_columns=["text_id"])))
```

```python
c.root.schema_for('Select by Metadata Attribute')
```

```python
n2 = c.root.add_subset_node(
    ("Select by Metadata Attribute",
     {'metadata_attribute': "text_id", 'value': "OMF"})
)
display(HTML(generate_concordance_html(c, n2, n=1000, metadata_columns=["text_id"])))
```

```python
display(HTML(generate_analysis_tree_html(c)))
```

```python
n2a = n2.add_arrangement_node(
    ordering = [
        ("Sort by Token-Level Attribute",
         {'tokens_attribute': "word", 'sorting_scope': "token", 'offset': -1})
    ]
)
display(HTML(generate_analysis_tree_html(c)))
display(HTML(generate_concordance_html(c, n2a, n=1000, metadata_columns=["text_id"])))
```

```python
n3 = c.root.add_arrangement_node(
    grouping = ("Partition by Ngrams",
         {'positions': [-2, -1]})
)
display(HTML(generate_analysis_tree_html(c)))
display(HTML(generate_concordance_html(c, n3, n=1000, metadata_columns=["text_id"])))
```

```python
n4 = n3.add_subset_node(
    ("Manual Line Selection",
     {'groups': ["('a', 'proud')"]})
)
display(HTML(generate_analysis_tree_html(c)))
display(HTML(generate_concordance_html(c, n4, n=1000, metadata_columns=["text_id"])))
```

```python
c.add_annotation(
    ("Annotate with Sentence Transformers",
     {'window_start': -10, 'window_end': 10, 'model_name': 'all-MiniLM-L6-v2'})
)
```

```python
n5 = c.root.add_arrangement_node(
    grouping = ("Flat Clustering by Embeddings",
                {"embeddings_column": "embeddings_sentence_transformers", "n_partitions": 5}),
    ordering = [
        ("Sort by Token-Level Attribute", {'offset': -1}),
        ("Sort by Token-Level Attribute", {'sorting_scope': "right"})
    ]
)
display(HTML(generate_concordance_html(c, n5, n=1000, metadata_columns=["text_id"])))
```

```python
display(HTML(generate_analysis_tree_html(c)))
```
