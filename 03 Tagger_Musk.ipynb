```python
import pandas as pd
import spacy

# Lade spaCy-Modell
nlp = spacy.load("en_core_web_sm")

# CSV-Datei einlesen
df = pd.read_csv("a_musk_quote_tweets.csv")

# Neue Spalten vorbereiten
all_tokens = []

print("POS-Tagging der Texte...")

for index, row in df.iterrows():
    doc = nlp(str(row['orig_tweet_text']))
    for token in doc:
        if not token.is_space:
            all_tokens.append({
                #"id": row["id"],
                "orig_tweet_id": row["orig_tweet_id"],
                "orig_tweet_created_at": row["orig_tweet_created_at"],
                "orig_tweet_text": row["orig_tweet_text"],
                "orig_tweet_url": row["orig_tweet_url"],
                "orig_tweet_twitter_url": row["orig_tweet_twitter_url"],
                "orig_tweet_username": row["orig_tweet_username"],
                "orig_tweet_retweet_count": row["orig_tweet_retweet_count"],
                "orig_tweet_reply_count": row["orig_tweet_reply_count"],
                "orig_tweet_like_count": row["orig_tweet_like_count"],
                "orig_tweet_quote_count": row["orig_tweet_quote_count"],
                "orig_tweet_view_count": row["orig_tweet_view_count"],
                "orig_tweet_bookmark_count": row["orig_tweet_bookmark_count"],
                "musk_tweet_id": row["musk_tweet_id"],
                "musk_quote_tweet": row["musk_quote_tweet"],
                "musk_quote_retweet_count": row["musk_quote_retweet_count"],
                "musk_quote_reply_count": row["musk_quote_reply_count"],
                "musk_quote_like_count": row["musk_quote_like_count"],
                "musk_quote_quote_count": row["musk_quote_quote_count"],
                "musk_quote_view_count": row["musk_quote_view_count"],
                "musk_quote_bookmark_count": row["musk_quote_bookmark_count"],
                "musk_quote_created_at": row["musk_quote_created_at"],
                "word": token.text,
                "lemma": token.lemma_,
                "pos": token.pos_,
                "lemma_p": f"{token.lemma_}_{token.pos_}"
            })

# Neues DataFrame mit Tokens
token_df = pd.DataFrame(all_tokens)

# Als neue CSV-Datei speichern
token_df.to_csv("a_musk_quote_tweets_tagged.csv", index=False, encoding="utf-8")

print("POS-Tagging abgeschlossen. Ergebnis gespeichert in 'a_musk_quote_tweets_tagged.csv'.")
```

```python
import asyncio
import re
import hashlib
import pandas as pd
from datetime import datetime
from playwright.async_api import async_playwright

# Hilfsfunktion: eindeutige ID pro Post
def make_post_id(data):
    key = f"{data['author']}|{data['date']}|{data['time']}|{data['text']}"
    return hashlib.md5(key.encode("utf-8")).hexdigest()

# Deine extract_metadata_text hier einfÃ¼gen!
def extract_metadata_text(text: str):
    lines = [line.strip() for line in text.strip().splitlines() if line.strip()]
    author_name = ""
    handle = ""
    platform = ""
    date_str = ""
    time_str = ""
    content_lines = []

    if len(lines) >= 2:
        author_name = lines[0].strip()
        match = re.search(
            r"(@[\w]+)\s*[â€¢\-]\s*(.*?)\s*[â€¢\-]\s*([A-Za-z]+ \d{1,2}, \d{4})\s*@\s*(\d{1,2}:\d{2} [AP]M)",
            lines[1]
        )
        if match:
            handle = match.group(1).strip()
            platform = match.group(2).strip()
            date_str = match.group(3).strip()
            time_str = match.group(4).strip()

        # Content ab Zeile 2 oder 3, ZeilenumbrÃ¼che durch Leerzeichen ersetzen
        start_idx = 2
        if len(lines) > 2 and lines[2].startswith("View"):
            start_idx = 3
        content_lines = lines[start_idx:]

    content_text = " ".join(content_lines).strip()

    try:
        dt = datetime.strptime(f"{date_str} {time_str}", "%B %d, %Y %I:%M %p")
        return {
            "author": f"{author_name} {handle}".strip(),
            "platform": platform,
            "date": dt.strftime("%Y-%m-%d"),
            "time": dt.strftime("%H:%M"),
            "year": int(dt.year),
            "month": dt.strftime("%B"),
            "day": int(dt.day),
            "text": content_text
        }
    except Exception:
        return {
            "author": f"{author_name} {handle}".strip(),
            "platform": platform,
            "date": date_str,
            "time": time_str,
            "year": "",
            "month": "",
            "day": "",
            "text": content_text
        }

async def scrape_all_dynamic(max_posts=100000, max_no_new=5):
    posts_data = []
    seen_ids = set()
    no_new_rounds = 0

    async with async_playwright() as p:
        browser = await p.firefox.launch(headless=True)
        page = await browser.new_page()
        await page.goto("https://rollcall.com/factbase-twitter/?platform=all&sort=date&sort_order=desc")

        while True:
            # auf neue Posts warten
            try:
                await page.wait_for_selector("div.block", timeout=15000)
            except:
                print("Timeout beim Warten auf Posts â€“ Abbruch.")
                break

            blocks = await page.query_selector_all("div.block")
            print(f"Aktuell {len(blocks)} Posts im DOM")

            new_in_round = 0
            for block in blocks:
                text = (await block.inner_text()).strip()
                data = extract_metadata_text(text)
                pid = make_post_id(data)

                if pid not in seen_ids:
                    seen_ids.add(pid)
                    posts_data.append(data)
                    new_in_round += 1

            print(f"Neue eindeutige Posts in dieser Runde: {new_in_round}")

            if new_in_round == 0:
                no_new_rounds += 1
                if no_new_rounds >= max_no_new:
                    print("Mehrfach keine neuen Posts gefunden â€“ beende Scraping.")
                    break
            else:
                no_new_rounds = 0

            print(f"Gesamt bisher gespeichert: {len(posts_data)}")

            if len(posts_data) >= max_posts:
                print("Maximale Anzahl Posts erreicht.")
                break

            # Scrollen zum Nachladen
            await page.evaluate("window.scrollBy(0, document.body.scrollHeight)")
            await page.wait_for_timeout(2000)

        await browser.close()

    df = pd.DataFrame(posts_data)
    df.to_csv("factbase_posts_clean.csv", index=False)
    print(f"Scraping abgeschlossen. Gesamt: {len(df)} eindeutige Posts gespeichert.")

# Falls du in Jupyter arbeitest:
# await scrape_all_dynamic()
```
