```python
import pandas as pd
import stanza
import re

stanza.download("en")

# Stanza-Pipeline mit Tweet-Tokenizer
nlp = stanza.Pipeline(
    lang="en",
    processors="tokenize,pos,lemma",
    tokenize_pretokenized=False,
    use_gpu=False,
    tokenize_with_spacy=False,
    tokenize_no_ssplit=False,
    tokenize_engine="tokenize/tweet"   # spezieller Tweet-Tokenizer
)


# Unicode-Emoji-Regex
emoji_pattern = re.compile("["
    u"\U0001F600-\U0001F64F"  # Emoticons
    u"\U0001F300-\U0001F5FF"  # Symbole & Piktogramme
    u"\U0001F680-\U0001F6FF"  # Transport & Symbole
    u"\U0001F1E0-\U0001F1FF"  # Flaggen
    "]+", flags=re.UNICODE)

# klassische Smileys
smiley_pattern = re.compile(r'[:;=8][\-~]?[)D]', flags=re.UNICODE)

def is_emoji_or_smiley(token):
    return bool(emoji_pattern.fullmatch(token)) or bool(smiley_pattern.fullmatch(token))


df = pd.read_csv("testkorpus_divers_50.csv")
all_results = []

for idx, row in df.iterrows():
    text = row.get("text")
    if pd.isna(text) or not isinstance(text, str) or text.strip() == "":
        continue

    doc = nlp(text)
    for sentence in doc.sentences:
        for word in sentence.words:
            xpos = word.xpos
            if is_emoji_or_smiley(word.text):
                xpos = "NFP"

            lemma_p = f"{word.lemma}_{word.upos}"
            all_results.append({
                "orig_tweet_id": row["orig_tweet_id"],
                "orig_tweet_created_at": row["orig_tweet_created_at"],
                "orig_tweet_text": row["orig_tweet_text"],
                "orig_tweet_url": row["orig_tweet_url"],
                "orig_tweet_twitter_url": row["orig_tweet_twitter_url"],
                "orig_tweet_username": row["orig_tweet_username"],
                "orig_tweet_retweet_count": row["orig_tweet_retweet_count"],
                "orig_tweet_reply_count": row["orig_tweet_reply_count"],
                "orig_tweet_like_count": row["orig_tweet_like_count"],
                "orig_tweet_quote_count": row["orig_tweet_quote_count"],
                "orig_tweet_view_count": row["orig_tweet_view_count"],
                "orig_tweet_bookmark_count": row["orig_tweet_bookmark_count"],
                "musk_tweet_id": row["musk_tweet_id"],
                "musk_quote_tweet": row["musk_quote_tweet"],
                "musk_quote_retweet_count": row["musk_quote_retweet_count"],
                "musk_quote_reply_count": row["musk_quote_reply_count"],
                "musk_quote_like_count": row["musk_quote_like_count"],
                "musk_quote_quote_count": row["musk_quote_quote_count"],
                "musk_quote_view_count": row["musk_quote_view_count"],
                "musk_quote_bookmark_count": row["musk_quote_bookmark_count"],
                "musk_quote_created_at": row["musk_quote_created_at"],
                "word": word.text,
                "lemma": word.lemma,
                "pos": word.upos,   
                "xpos": xpos,
                "lemma_p": lemma_p,
            })

pos_df = pd.DataFrame(all_results)
output_file = "testkorpus_divers_50_stanza_tweets2.csv"
pos_df.to_csv(output_file, index=False, encoding="utf-8")
display(pos_df.head(20))
```
