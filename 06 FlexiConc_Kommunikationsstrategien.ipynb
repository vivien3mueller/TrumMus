#### 11. Autorit√§tsargument (Appeal to authority): Wahrheitsgehalt durch Autorit√§t begr√ºnden
- I, because, Elon, engineer, scientist, expert

```python
# [lemma="I"]
```

```python
i = Concordance()
i.retrieve_from_cwb( 
    registry_dir="/usr/local/share/cwb/registry",
    corpus_name="TRUMP",
    query='[lemma="I"]',
    tokens_attrs=["word", "lemma", "pos", "x_pos", "lemma_p"],
    metadata_attrs=["text_id", "text_date", "text_time", "text_day", "text_month", "text_year", "text_favorites", "text_retweets", "text_isRetweet", "text_isDeleted", "text_device", "text_isFlagged", "text_platform"],
    context=20
)
display(HTML(generate_analysis_tree_html(i)))
display(HTML(generate_concordance_html(i, i.root, n=1000, metadata_columns=["text_date"])))
```

```python
i1 = i.root.add_subste_node(
    ("random Sample",
    {'sample_size': 20, 'seed': 45})
)
display(HTML(generate_concordance_html(i, i1, n=1000)))
```

```python
i2 = i.root.add_arrangement_node(
    grouping = ("Partition by Ngrams",
         {'positions': [-2, -1]})
)
display(HTML(generate_analysis_tree_html(i)))
display(HTML(generate_concordance_html(i, i2, n=1000, metadata_columns=["text_date"])))
```

```python
i3 = i.root.add_arrangement_node(
    grouping = ("Partition by Ngrams",
         {'positions': [1, 2]})
)
display(HTML(generate_analysis_tree_html(i)))
display(HTML(generate_concordance_html(i, i3, n=1000, metadata_columns=["text_date"])))
```

```python
i4 = i.root.add_arrangement_node(
    grouping = ("Partition by Metadata Attribute",
         {'metadata_attribute': 'text_year'})
)
display(HTML(generate_analysis_tree_html(i)))
display(HTML(generate_concordance_html(i, i4, n=1000, metadata_columns=["text_date"])))
```
