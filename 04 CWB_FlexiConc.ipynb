Die zu verarbeitenden Dateien m√ºssen als VRT vorliegen, sodass sie in CWB eingelesen werden k√∂nnen. FlexiConc arbeitet mit CWB, sodass die Daten gut verarbeitet werden k√∂nnen. Zun√§chst wird nun das Korpus MUSK, anschlie√üend das Korpus TRUMP eingelesen. Im Anschluss daran wird FlexiConc installiert und getestet.

Maximum f√ºr CWB: 3,4 billion token

```python
# Wie viele Token haben die beiden Dateien?
from pathlib import Path

def count_tokens(vrt_path):
    token_count = 0
    with open(vrt_path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            # √ºberspringe Struktur-Tags
            if not line or line.startswith("<") and line.endswith(">"):
                continue
            token_count += 1
    return token_count

# Pfade zu den VRT-Dateien
musk_vrt = Path("/usr/local/share/cwb/data/musk/musk.vrt")
trump_vrt = Path("/usr/local/share/cwb/data/trump/trump.vrt")

musk_tokens = count_tokens(musk_vrt)
trump_tokens = count_tokens(trump_vrt)

print(f"Anzahl der Token in Musk-Korpus: {musk_tokens}")
print(f"Anzahl der Token in Trump-Korpus: {trump_tokens}")
```
