#### MUSK

```python
cwb-encode -d /usr/local/share/cwb/data/musk \
           -xsBC9 -c ascii -f musk.vrt \
           -R /usr/local/share/cwb/registry/musk \
           -P lemma -P pos -P x_pos -P lemma_p \
           -S s -S text \
           -S id -S url -S twitterUrl \
           -S retweetCount -S replyCount -S likeCount \
           -S quoteCount -S viewCount -S createdAt \
           -S bookmarkCount -S isReply -S inReplyToId \
           -S conversationId -S inReplyToUserId -S inReplyToUsername \
           -S isPinned -S isRetweet -S isQuote \
           -S isConversationControlled -S possiblySensitive \
           -S quoteId -S quote -S retweet \
           -S day -S month -S year -S time
# eher so
```

```python
# Indexieren
$ cwb-make -V MUSK
```

```python
echo $CORPUS_REGISTRY
```

```python
# Überprüfen
cwb-describe-corpus MUSK -s
```

#### TRUMP

```python
$ cwb-encode -d /usr/local/share/cwb/data/trump \
    -xsBC9 -c ascii \
    -f trump.vrt \
    -R /usr/local/share/cwb/registry/trump \
    -P lemma -P pos -P x_pos -P lemma_p \
    -S s -S text \
    -S id -S date -S time -S day -S month -S year \
    -S favorites -S retweets -S isRetweet -S isDeleted \
    -S device -S isFlagged -S platform
# in Zeile 3: /usr/local/share/cwb/registry/trump/trump.vrt
```

```python
# Indexieren
$ cwb-make -V TRUMP
```

```python
echo $CORPUS_REGISTRY
```

```python
# Überprüfen
cwb-describe-corpus TRUMP -s
```
